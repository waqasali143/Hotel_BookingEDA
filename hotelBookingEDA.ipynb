adr mean Average Daily Rate

```python
df['adr'].plot(kind='box')
```

```python
# Now we remove the outliers
df = df[df['adr']< 5000]
```

```python
canceled_reservation_inPercentage = df['is_canceled'].value_counts(normalize=True)*100
print(canceled_reservation_inPercentage)

# 0 mean not conceled and 1 mean canceled
```

```python
plt.figure(figsize=(5, 4))
plt.title('Reservation Status')
plt.bar(['Not conceled', 'Conceled'], df['is_canceled'].value_counts())
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x = 'hotel', hue= 'is_canceled', data=df, palette='Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in Different Hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('Number of Reservation')
plt.legend(['Not conceled', 'Conceled'])
plt.show()
```

### Check the Reservation Status Resort Hotel and City Hotel in %

Resort Hotel Status 

```python
# 0 mean Not conceled and 1 mean conceled
Resort_hotel = df[df['hotel'] == 'Resort Hotel']
Resort_hotel['is_canceled'].value_counts(normalize=True)*100
```

City Hotel Status

```python
# 0 mean Not conceled and 1 mean conceled
City_hotel = df[df['hotel'] == 'City Hotel']
City_hotel['is_canceled'].value_counts(normalize=True)*100
```

City hotel reservation cancellation rate is too much

```python
# adr mean Average Daily Rate
Resort_hotel = Resort_hotel.groupby('reservation_status_date')[['adr']].mean()
City_hotel = City_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(Resort_hotel.index, Resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(City_hotel.index, City_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax2 = sns.countplot(x = 'month', hue='is_canceled', data=df, palette='bright')
legend_labels,_ = ax2. get_legend_handles_labels()
ax2.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status Per Month', size = 20)
plt.xlabel('month')
plt.ylabel('Number of Reservation')
plt.legend(['Not canceled', 'Canceled'])
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title('ADR Per Month', fontsize=30)
sns.barplot(data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(), x='month', y='adr')
plt.show()         
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top10_country = cancelled_data['country'].value_counts()[:10]

plt.figure(figsize=(8,8))
plt.title('Top 10 Country with Reservation canceled')
plt.pie(top10_country, autopct= '%.2f', labels=top10_country.index)
plt.show()
```

```python
df['market_segment'].value_counts()
```

```python
# market segment values in percentage
df['market_segment'].value_counts(normalize=True)*100
```

```python
# Canceled data result
cancelled_data['market_segment'].value_counts(normalize=True)*100
```

```python
canceled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_canceled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date', inplace=True)
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'], label = 'Not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'], label = 'Canceled')
plt.legend()
```

### Draw the chart of the Average Daily Rate

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'], label = 'Not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'], label = 'Canceled')
plt.legend(fontsize = 20)
plt.show()
```
